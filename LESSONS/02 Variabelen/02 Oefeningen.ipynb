#### Oefening 4: Input

```python
getallen = input('Geef 3 getallen, gescheiden door ; ')

getal_1 = int(getallen.split(';')[0])
getal_2 = int(getallen.split(';')[1])
getal_3 = int(getallen.split(';')[2])

print(f"Gemiddelde van de getallen is {(getal_1 + getal_2 + getal_3)/3}")
```
